```python
for result in results:
    nyt_scrape['school'].append(result.find("div", class_='schoolname').find('h2').text)
    nyt_scrape['cases'].append(result.find("div", class_='schoolcases').find('p').text)
    nyt_scrape['location'].append(result.find("div", class_='schoolloc').find('p').text)
    nyt_scrape['school_logo'].append(result.find("div", class_='logo').find('img')['src'])
```

```python
nyt_scrape
```

```python
nyt_scrape_df = pd.DataFrame(nyt_scrape)
```

```python
nyt_scrape_df
```

```python
for i in range(len(nyt_scrape_df['school'])):
    if ('at' in nyt_scrape_df.loc[i,'school']):
        nyt_scrape_df.loc[i, 'school'] = nyt_scrape_df.loc[i, 'school'].replace(' at ', ' - ')
    if (', ' in nyt_scrape_df.loc[i,'school']):
        nyt_scrape_df.loc[i, 'school'] = nyt_scrape_df.loc[i, 'school'].replace(', ', ' - ')
        
    if(nyt_scrape_df.loc[i, 'school'][-1:] == "*"):
        nyt_scrape_df.loc[i, 'school']  = nyt_scrape_df.loc[i, 'school'].replace("*", "")
        
```

```python
for i in range(len(nyt_scrape_df['school'])):
    if ('-' in nyt_scrape_df.loc[i,'school']):
        #nyt_scrape_df.loc[i, 'school'] = nyt_scrape_df.loc[i, 'school'].replace(' at ', ' - ')
        dash_loc = nyt_scrape_df.loc[i, 'school'].find('-')
        if nyt_scrape_df.loc[i, 'school'][dash_loc-1].islower():
            nyt_scrape_df.loc[i, 'school'] = nyt_scrape_df.loc[i, 'school'].replace('-', ' - ')
```
