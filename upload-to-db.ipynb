```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
pop_csv= ("county_state_population.csv")
pop_data_df = pd.read_csv(pop_csv, index_col=0)
pop_data_df.index.names = ['countystateid']
pop_data_df.rename(columns={"population 2019": "population_2019", "county-state": "county_state"}, inplace=True)
pop_data_df.head()
```

```python
covid_csv= (r"C:\Users\User\Desktop\etl_project\covid_county_test_data_7-28-2020.csv")
covid_df = pd.read_csv(covid_csv,index_col=0)
covid_df.index.names = ['covidid']
covid_df.rename(columns={"county-state": "county_state"}, inplace=True)
covid_df.head()
```

```python
ranked_schools_csv= ("ranked_schools_with_county.csv")
ranked_schools_df = pd.read_csv(ranked_schools_csv,index_col=0)
ranked_schools_df.index.names = ['rankedschoolsid']
ranked_schools_df.rename(columns={"College Name": "college_name", "Rank": "rank", "Number of Full-Time Undergraduates": "number_of_full_time_undergraduates", "Niche Overall Grade":"niche_overall_grade", "Location": "location","Acceptance Rate":"acceptance_rate"}, inplace=True)
ranked_schools_df["number_of_full_time_undergraduates"] = ranked_schools_df["number_of_full_time_undergraduates"].str.replace(",","").astype(int)
ranked_schools_df["acceptance_rate"] = ranked_schools_df["acceptance_rate"].str.replace("%","").astype(float)
ranked_schools_df.head()
```

```python
school_cases_csv= ("school_cases.csv")
school_cases_df = pd.read_csv(school_cases_csv,index_col=0)
school_cases_df.index.names = ['schoolid']
school_cases_df.rename(columns={"College Name": "college_name"}, inplace=True)
school_cases_df.head()
```

```python
school_county_csv= ("school_county_list.csv")
school_county_df = pd.read_csv(school_county_csv, index_col=0)
school_county_df.rename(columns={ "College Name": "college_name", "Location": "location"}, inplace=True)
school_county_df.index.names = ['county_state']
school_county_df.head(100)
```

```python
connection_string = "postgres:postgres@localhost:5432/etl"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
 engine.table_names()
```

```python
ranked_schools_df.to_sql(name='rankedschoolswithcounty', con=engine, if_exists='append', index=True)
```

```python
pop_data_df.to_sql(name='countystatepopulation', con=engine, if_exists='append', index=True)
```

```python
school_county_df.to_sql(name='schoolcounty', con=engine, if_exists='append', index=True)
```

```python
school_cases_df.to_sql(name='schoolcases', con=engine, if_exists='append', index=True)
```

```python
covid_df.to_sql(name='covidcountytestdata_7_28_20', con=engine, if_exists='append', index=True)
```
