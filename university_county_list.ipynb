```python
school_dictionary = {}
```

```python
school_dictionary = {'Rank': [],'College Name': [], 'address': [], 'Number of Full-Time Undergraduates': [], 'Niche Overall Grade' : [], 'Location': [], 'Acceptance Rate': [], 'niche_url': []}
```

```python
for result in results:
    print(result)
    try:
        print("__________")

        value_list = result.find_all('span', class_='td-value')
        print(value_list)
        school_dictionary['Rank'].append(value_list[0].text)
        school_dictionary['College Name'].append(value_list[1].text)
        school_dictionary['Number of Full-Time Undergraduates'].append(value_list[2].text)
        school_dictionary['Niche Overall Grade'].append(value_list[3].text)
        school_dictionary['Location'].append(value_list[4].text)
        school_dictionary['Acceptance Rate'].append(value_list[5].text)
        school_dictionary['niche_url'].append(value_list[1].find('a')['href'])
    except:
        print("")
```

```python
for x in range(len(school_dictionary['address'])):
    url = 'https://www.whatcountyamiin.com/'
    browser.visit(url)
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    browser.find_by_id('address').fill(school_dictionary['address'][x])
    browser.find_by_id('center').click()
    time.sleep(2)
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    results = soup.find_all(id = 'location')
    try:
        county = results[0].find_all('p')[0].find('span').find('strong').text
        print(county)
        school_dictionary['county'].append(county)
    except:
        school_dictionary['county'].append('error')
```
