```python
import pandas as pd
import numpy as np
import us
```

```python
input_csv = 'ranked_schools_with_county.csv'
output_csv = 'school_county_list.csv'
nyt_scrape = 'nyt_scrape.csv'
```

```python
school_county_df = pd.read_csv(input_csv)
```

```python
school_county_df.columns
```

```python
school_county_df
```

```python
school_county_df = school_county_df.drop(columns = ['address', 'niche_url', 'Number of Full-Time Undergraduates', 'Rank', 'Unnamed: 0', 'Niche Overall Grade', 'Acceptance Rate'])
```

```python
school_county_df
```

```python
for i in range(len(school_county_df['College Name'])):
    school_county_df.loc[i, 'state'] = us.states.lookup(school_county_df.loc[i, 'Location'][-2:]).name
```

```python
school_county_df
```

```python
school_county_df['county-state'] = school_county_df['county'] + ' - ' + school_county_df['state']
```

```python
school_county_df.set_index('county-state', inplace = True)
```

```python
school_county_df.to_csv(output_csv)
```

```python
nyt_scrape_df = pd.read_csv(nyt_scrape)
```
